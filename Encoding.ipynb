#### when y is categorical data then we use Label Encoder 
- so for X categorical data ordinal Encoding and for y we use lable encoder

```python
customer=pd.read_csv("customer.csv")
```

```python
customer.sample(5)
```

```python
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import OrdinalEncoder
```

```python
oe=OrdinalEncoder(categories=[["Poor","Average","Good"],['School','UG','PG']])
```

```python
x=customer[["review","education"]]
y=customer[['purchased']]
```

```python
from sklearn.model_selection import train_test_split
```

```python
oe.fit(x)
x_scaled = oe.transform(x)
le = LabelEncoder()
le.fit(y)
y_scaled = le.fit_transform(y)
```

```python
X_train,X_test,y_train,y_test=train_test_split(x_scaled,y_scaled,test_size=0.2,random_state=0)
```
